```python
import scanpy as sc
import pandas as pd
import numpy as np
```

```python
#load rnk files
c2=pd.read_csv("c2_gsea/report_pos.tsv",sep="\t")
c1=pd.read_csv("c1_gsea/report_pos.tsv",sep="\t")
```

```python
#plot GSEA results with NES on X axis and colored by FDR
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8,13))
dpl=plt.scatter(c2["NES"][0:25],np.flip([i for i in range(285)])[0:25],s=400,c=c2["FDR q-val"][0:25],cmap='cool_r',vmin=0,vmax=1)
plt.yticks(ticks=np.flip([i for i in range(285)])[0:25],labels=c2["NAME"][0:25])
plt.xlabel("Normalized Enrichment Score")
cbar=plt.colorbar(dpl,fraction=0.05)
plt.title("GSEA Results for Clone 2")
plt.savefig("supp_7d.pdf",format='pdf',bbox_inches='tight')
```
